Try to join the actor, film_actor, and film table together (3 joins!)

```python
SELECT * FROM film AS film
FULL JOIN
film_actor as film_actor
ON film.film_id = film_actor.film_id
FULL JOIN 
actor AS actor
ON film_actor.last_update = actor.last_update;
```
